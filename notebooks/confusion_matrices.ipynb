```python
import pickle
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
sns.set()
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../data/tags.pickle','rb') as rf:
    tags = pickle.load(rf)

with open('../data/confusion_matrices.pickle','rb') as rf:
    confusions = pickle.load(rf)
```

```python
list(enumerate(tags))
```

```python
confusions.shape
```

```python
plt.subplots(nrows=2,ncols=2,figsize=(24,16))

for ind, val in enumerate([0,8,9,13]):
    plt.subplot(2,2,ind+1)
    sns.heatmap(confusions[val,:,:],annot=True,fmt='.0f',cmap='BuGn',cbar=False,annot_kws={'size': 35},)
    plt.xlabel('Prediction',fontsize=35)
    plt.ylabel('True Value',fontsize=35)
    plt.title(tags[val].title(),fontsize=50)
    plt.xticks(fontsize=25)
    plt.yticks(fontsize=25)
    plt.tight_layout()

plt.savefig('../visuals/confusion.png',dpi=100)
```
