```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('../data/train_v2.csv')
```

```python
cv = CountVectorizer()
cv_matrix = cv.fit_transform(df.tags)
cv_df = pd.DataFrame(cv_matrix.toarray(),columns=cv.get_feature_names())
```

```python
cv_df.head()
```

```python
# cv_df.to_csv('../data/count_vectorizer.csv',index=False)
```

```python
co_matrix = np.matmul(cv_df.T.values,cv_df.values)
co_zero_diag = np.matmul(cv_df.T.values,cv_df.values)
np.fill_diagonal(co_zero_diag,0)

co_occurence = pd.DataFrame(co_matrix,
                           columns = cv_df.columns,
                           index=cv_df.columns)

co_zero = pd.DataFrame(co_zero_diag,
                      columns=cv_df.columns,
                      index=cv_df.columns)
```
